```python
x=int(input("enter 1st number"))
y=int(input("enter 2nd number"))
def calc(x,y):
    z=input("enter operation")
    if z=="+":
       print(x+y)
    elif z=="-":
       print(x-y)
    elif z=="*":
       print(x*y)
    elif z=="/" and y==0:
       print("error")
    elif z=="/":
       print(x/y)
    elif z=="%":
       print(x%y)
    elif z=="//":
       print(x//y)
    elif z=="^":
       print(x**y)
    else:
       print("error")
calc(x,y)
```
